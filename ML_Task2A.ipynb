```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r'D:\semester 10\[NETW1013] Machine Learning\Lab 2\GUC_2350_68_65247_2026-02-25T11_06_51\CC GENERAL.csv')

# numeric vs categorical features
print("--- Dataset Info ---")
df.info()

# distributions of numeric features (min, max, mean, std)
print("\n--- Numeric Feature Summary ---")
display(df.describe())

# skewness
print("\n--- Skewness ---")
display(df.skew(numeric_only=True))

# missing values
print("\n--- Missing Values ---")
display(df.isnull().sum())

# correlations between numeric features
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm", annot_kws={"size": 8})
plt.title("Correlation Matrix")
plt.show()

# outliers using boxplots
plt.figure(figsize=(16, 8))
sns.boxplot(data=df.drop(columns=['CUST_ID']))
plt.xticks(rotation=90)
plt.title("Boxplots for Outlier Detection")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

df_prepared = df.copy()

# CUST_ID is just a unique text identifier, it holds no mathematical value for clustering
if 'CUST_ID' in df_prepared.columns:
    df_prepared = df_prepared.drop(columns=['CUST_ID'])

# missing values
# we fill them with the median value of their respective columns.
# We use median instead of mean because credit card data has heavy outliers.
imputer = SimpleImputer(strategy='median')
df_imputed = pd.DataFrame(imputer.fit_transform(df_prepared), columns=df_prepared.columns)

print("Missing values after imputation:")
print(df_imputed.isnull().sum().sum())

# Scale numeric features
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_imputed), columns=df_imputed.columns)

# final dataset
print("\n--- Final Dataset Verification ---")
print(f"Dataset shape: {df_scaled.shape}")
display(df_scaled.head())
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 6))
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")

# Calculate the linkage matrix and plot the dendrogram
linkage_matrix = shc.linkage(df_scaled, method='ward')
dend = shc.dendrogram(linkage_matrix, truncate_mode='lastp', p=30, show_leaf_counts=True)
```

```python
from sklearn.cluster import AgglomerativeClustering

# Create the model using 7 clusters, euclidean distance, and ward linkage
agg_model = AgglomerativeClustering(n_clusters=7, metric='euclidean', linkage='ward')

# Fit the model and predict the cluster labels
cluster_labels = agg_model.fit_predict(df_scaled)

# Attach the predicted labels to our unscaled dataset so we can interpret them later
df_prepared['Cluster'] = cluster_labels
print("Model fitted and cluster labels assigned successfully!")

# how many customers were placed into each cluster
print("Customers per cluster:")
print(df_prepared['Cluster'].value_counts())
```

```python
from sklearn.decomposition import PCA
import seaborn as sns

# Reduce dimensions to 2D
pca = PCA(n_components=2)
pca_features = pca.fit_transform(df_scaled)

# how the clusters separate based on 2 principal components
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_features[:, 0], y=pca_features[:, 1], hue=cluster_labels, palette='viridis', alpha=0.6)
plt.title('Agglomerative Clustering Results (2D PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
# how the clusters separate based on 2 real metrics
# BALANCE vs PURCHASES
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_prepared, x='BALANCE', y='PURCHASES', hue='Cluster', palette='viridis', alpha=0.6)
plt.title('Customer Balance vs. Purchases by Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Calculate Silhouette Score
sil_score = silhouette_score(df_scaled, cluster_labels)
print(f"Silhouette Score for 7 clusters: {sil_score:.4f}")

# Calculate the mean values for each feature grouped by the cluster
# This tells us the behavior of the customers in each group
cluster_profile = df_prepared.groupby('Cluster').mean()
display(cluster_profile)
```

```python


# Based on the evaluation metrics and visualizations, the clustering quality is fair, but exhibits heavy overlapping due to the nature of the dataset.

# Justification:

# Silhouette Score:
#The model achieved a Silhouette Score of 0.1617
#because this value is close to 0, it indicates that the clusters are not entirely distinct
#and have overlapping boundaries in the multi-dimensional space.

# Impact of Outliers:
#Credit card datasets naturally contain extreme outliers (a few customers spending 49000 while the median is 361)
#Agglomerative Clustering is highly sensitive to these distances
#Instead of grouping these "whales" with normal users, the algorithm successfully isolated them into small micro-clusters (a cluster of just 23 people)

# PCA Visualization:
#The 2D PCA plot confirms this behavior
#The majority of regular customers form a dense overlapping central mass, while the extreme spenders pull away into their own distinct outer groups.
```
