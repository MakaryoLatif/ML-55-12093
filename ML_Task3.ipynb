```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv(r'D:\semester 10\[NETW1013] Machine Learning\Lab 3\GUC_2350_68_65768_2026-03-04T17_32_48\Lab 3\sonar.all-data.csv')

# Split data into features (X) and labels (y)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Split into a training set and test set
# test_size=0.10 sets 10% for the final test set 90% for cross-validation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline

# StandardScaler and KNN model
scaler = StandardScaler()
knn = KNeighborsClassifier()

# pipeline operations as a list of tuples: ('string_name', object)
operations = [('scaler', scaler), ('knn', knn)]
pipe = Pipeline(operations)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to test
k_values = list(range(1, 30))
param_grid = {'knn__n_neighbors': k_values,
              'knn__weights': ['uniform', 'distance'],
              'knn__metric': ['euclidean', 'manhattan', 'minkowski']
             }
 
# cv=5 means 5-fold cross-validation
grid_classifier = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy')

# Fit the grid search to the training data
grid_classifier.fit(X_train, y_train)

# Report back the best performing parameters
print("Best Parameters:", grid_classifier.best_params_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# predict the labels for our test set
y_pred = grid_classifier.predict(X_test)

# Print out the confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\n")

# Print out the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
### Final Model Conclusion & Evaluation

# Using a Pipeline (StandardScaler + KNeighborsClassifier) and `GridSearchCV`, we successfully trained a model to differentiate between
#submerged rocks and naval mines based on 60 sonar frequency responses

# Best Parameters:'metric': 'euclidean', 'n_neighbors': 1, 'weights': 'uniform'
# While `k=1` typically signals overfitting, our test data evaluation proves otherwise
# The boundary between a rock's sonar echo and a mine's echo in this 60-dimensional space is highly complex
# For this specific dataset relying strictly on the single closest neighbor yields the most accurate predictions without adding noise

# The final model achieved  90% overall accuracy on unseen test data
# Out of 21 test samples
#the model correctly identified 12 mines and 7 rocks
#It only made 2 errors: 1 false mine (classified a rock as a mine) and 1 false rock (classified a mine as a rock)
```
